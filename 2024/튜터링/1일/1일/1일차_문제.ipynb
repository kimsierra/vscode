### 11. '회사명', '거래소코드', '회계년도', '영업이익' 으로만 이루어진 데이터프레임 생성하기
- 만들어진 데이터프레임을 df_2로 저장하기
- df_2의 회계년도 열을 datetime형식으로 변환하기
- 회계년도의 연도값을 year이라는 새로운 열에 저장하기
- year값이 23년이면서, 영업이익이 적자인 회사명 출력

```python
df_2 = df[['회사명', '거래소코드', '회계년도', '영업이익']]
```

```python
# df_2의 회계년도 열을 datetime형식으로 변환하기
df_2['회계년도']= pd.to_datetime(df_2['회계년도'])
```

```python
# 회계년도의 연도값을 year이라는 새로운 열에 저장하기
df_2['year']= df_2['회계년도'].dt.year
df_2
```

```python
# year값이 23년이면서, 영업이익이 적자인 회사명 출력
df_2[(df_2['year']==2023) & (df_2['영업이익'] < 0)]
# df_2['year'] == 2023
```
