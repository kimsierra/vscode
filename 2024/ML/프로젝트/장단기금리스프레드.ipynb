```python
import pandas as pd

df = pd.read_excel("C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트/data/장단기금리스프레드_PRE.xlsx")
df.drop([0,1,2,3,4,5],inplace=True)
df.rename(columns={'Unnamed: 1':'국고채(3년) - 기준금리(%p)',
                   'Unnamed: 2':'국고채(3년) - 통화안정증권(91일)(%p)',
                   'Unnamed: 3':'국고채(10년) - 국고채(3년)(%p)',
                   '장단기 금리 스프레드':'기간'}, inplace=True)
df
```

```python
df['기간'] = pd.to_datetime(df['기간'])
df
```

```python
df = df.groupby(df['기간'].dt.to_period('M')).mean().reset_index(drop=True)
df
```

```python
df.info()
```

```python
df['증감_국고채(3년)_기준금리'] = (df['국고채(3년) - 기준금리(%p)'] - df['국고채(3년) - 기준금리(%p)'].shift(1)) / df['국고채(3년) - 기준금리(%p)'].shift(1)
df['증감_국고채(3년)_통화안정증권(91일)'] = (df['국고채(3년) - 통화안정증권(91일)(%p)'] - df['국고채(3년) - 통화안정증권(91일)(%p)'].shift(1)) / df['국고채(3년) - 통화안정증권(91일)(%p)'].shift(1)
df['증감_국고채(10년)_국고채(3년)'] = (df['국고채(10년) - 국고채(3년)(%p)'] - df['국고채(10년) - 국고채(3년)(%p)'].shift(1)) / df['국고채(10년) - 국고채(3년)(%p)'].shift(1)
df['기간'] = df['기간'].dt.strftime('%Y-%m')
```

```python
df.to_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트/data/장단기금리스프레드_PRE2.xlsx')
```
