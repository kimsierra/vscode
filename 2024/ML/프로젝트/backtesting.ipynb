```python
### 가격 데이터 불러오기
kospi = fdr.DataReader("KS11","20210101","20221231")
kospi
```

```python
# 1. index 초기화
kospi.reset_index(inplace=True)
kospi
```

```python
# 월말 데이터로 변경하기 위한 Group by 이용 데이터 수정
how = {
    'Open' : 'first',
    'High' : 'max',
    'Low' : 'min',
    'Close' : 'last',
    'Volume' : 'sum'

}
kospi_month = kospi.groupby(pd.Grouper(key='Date',freq='m')).agg(how)
kospi_month
```

```python
# 월마다의 수익률 계산을 위해 월별 시초와 종가의 수익률을 확인
kospi_month["Change"] = (kospi_month["Close"] - kospi_month["Open"])/kospi_month["Open"] *100
kospi_month
```

```python
# 1. index 초기화
kospi_month.reset_index(inplace=True)
kospi_month
```

```python
# 날짜의 데이터 타입 확인
# 만약 object인경우 pd.to_datetime을 통해 데이터 타입 변경
kospi_month.info()
```

```python
#파생변수로 날짜의 월을 만들기
kospi_month['month'] = kospi_month['Date'].dt.month
kospi_month
```

```python
kospi_month.set_index("Date",inplace=True)
kospi_month
```

```python
# groupy 를 통해 평균과 표준편차 구하기, 및 DataFrame 화 연율화시켜서 확인
change_mean = kospi_month.groupby(by='month')["Change"].mean()
change_std= kospi_month.groupby(by='month')["Change"].std()

print(change_mean.sort_values(ascending=False))
print(change_std.sort_values(ascending=False))
```

```python
# 각 월별 평균과 표준편차 데이터 프레임 화
month_return = pd.DataFrame( data= {'month':change_mean.index,
                            'mean': change_mean,
                            'std':change_std})
month_return
```

```python
import bt

# 데이터 불러오기
data = bt.get('spy, agg', start='2010-01-01', end='2022-12-31')

# 전략 생성 함수
def create_strategy(data):
    
    # 조건에 따라 가중치를 동적으로 변경하는 사용자 정의 알고리즘
    class ConditionBasedWeights(bt.Algo):
        def __call__(self, target):
            month = target.now.month
            if month % 2 == 0:
                weights = {'spy': 0.6, 'agg': 0.4}
            else:
                weights = {'spy': 0.4, 'agg': 0.6}
            target.temp['weights'] = weights
            return True

    # 전략 구성
    strategy = bt.Strategy('Economic_Condition_Based',
        [
            bt.algos.RunMonthly(),
            bt.algos.SelectAll(),
            ConditionBasedWeights(),
            bt.algos.Rebalance()
        ]
    )
    return strategy

# 전략 생성 및 백테스트 실행
strategy = create_strategy(data)
backtest = bt.Backtest(strategy, data)
result = bt.run(backtest)

# 결과 출력
result.plot(title='Economic Condition Based Portfolio Backtest')
result.display()
```

```python
import bt
import pandas as pd

# 데이터를 불러온 후 NaN 값 처리
stock = bt.get('069500.KS', start='2010-01-01')  # KODEX 200
bond = bt.get('152380.KS', start='2010-01-01')  # KODEX 국고채 10년


# 필요한 경우 데이터프레임을 다시 결합
data = pd.concat([stock, bond], axis=1)
data.columns = ['KODEX 200', 'KODEX 국고채 10년']

# 사용자 정의 기간 입력
expansion_periods = [('2013-01-01', '2015-01-01'), ('2017-01-01', '2019-01-01')]  # 확장국면 기간 예시
contraction_periods = [('2015-01-01', '2017-01-01'), ('2019-01-01', '2021-01-01')]  # 수축국면 기간 예시

# 전략 생성 함수
def create_strategy(data, expansion_periods, contraction_periods):
    # 가중치 조정 알고리즘
    class AdjustWeights(bt.Algo):
        def __call__(self, target):
            current_date = target.now
            for start, end in expansion_periods:
                if pd.to_datetime(start) <= current_date <= pd.to_datetime(end):
                    target.temp['weights'] = {'KODEX 200': 0.6, 'KODEX 국고채 10년': 0.4}
                    return True
            for start, end in contraction_periods:
                if pd.to_datetime(start) <= current_date <= pd.to_datetime(end):
                    target.temp['weights'] = {'KODEX 200': 0.4, 'KODEX 국고채 10년': 0.6}
                    return True
            return False

    # 전략 구성
    strategy = bt.Strategy('Economic_Cycle_Adjusted',
        [
            bt.algos.RunDaily(),
            bt.algos.SelectAll(),
            AdjustWeights(),
            bt.algos.Rebalance()
        ]
    )
    
    return strategy

# 전략 생성 및 백테스트 실행
strategy = create_strategy(data, expansion_periods, contraction_periods)
backtest = bt.Backtest(strategy, data)
result = bt.run(backtest)

# 결과 출력
result.plot(title='Economic Cycle Adjusted Portfolio Backtest')
result.display()
```
