```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=112)

dt_clf_3 = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_clf_3.fit(X_train, y_train)
pred = dt_clf_3.predict(X_test)
accuracy_score(y_test, pred)
```

```python
dt_clf_3.classes_
```

```python
dt_clf_3.feature_importances_ # 피쳐중요도(꽃밭침 길이, 폭, 꽃잎의 길이, 폭)-결정트리가 분기할 때 꽃잎의 길이 피쳐를 많이 
                            # 사용함, 어떤 피쳐를 선택해야 하는지 (정보획득량이 많은 쪽으로 분기)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=112)

dt_clf = DecisionTreeClassifier(random_state=42)
dt_clf.fit(X_train, y_train)
pred = dt_clf.predict(X_test)
accuracy_score(y_test, pred)
```

```python
dt_clf.classes_
```

```python
dt_clf.feature_importances_
```
