```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("코스피_코스닥_결합_결측치.csv", index_col=0)
df1
```

```python
df2 = pd.read_csv("시장지표.csv", index_col=0, encoding='cp949')
df2
```

```python
df2.columns
```

```python
df1.columns
```

```python
df = df1[['회사명','회계년도','부실']]
df
```

```python
df1 = df1.drop(columns='부실')
df1.columns
```

```python
# 1. 재무비율 데이터프레임에서 연도 추출
# accounting_year 열을 문자열로 변환
df1['회계년도'] = df1['회계년도'].astype(str)

# 연도 부분만 추출하여 새로운 'year' 열을 생성
df1['회계년도'] = df1['회계년도'].str[:4].astype(int)

df['회계년도'] = df['회계년도'].str[:4].astype(int)
# 2. 두 데이터프레임 결합
merged_df = pd.merge(df1, df2, on='회계년도', how='left')

merged_df
```

```python
result = pd.merge(merged_df, df, on=['회사명','회계년도'], how='left')
```

```python
result
```

```python
result.to_csv("코스피_코스닥_시장지표_결합.csv")
```
