```python
import pandas as pd

# 파일 경로 설정
file_path_all_features = 'C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/data/연결_개별_합(코스피).csv'
file_path_HS = 'C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/data/코스피개별연결피쳐(최종2).xlsx'
file_path_HK = 'C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/data/코스피효경개별연결피쳐(최종2).xlsx'
file_path_HM = 'C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/data/코스피_해미피쳐_연결_개별_수정.csv'

# CSV 파일 불러오기
all_features = pd.read_csv(file_path_all_features, index_col=0)

# 엑셀 파일 불러오기
HS_features = pd.read_excel(file_path_HS)
HK_features = pd.read_excel(file_path_HK)

# CSV 파일 불러오기
HM_features = pd.read_csv(file_path_HM, index_col=0)

# 필요한 컬럼들 선택
HS_columns = ['매출액증가율', '총자본증가율', '매출액총이익률', '노동소득분배율', '재고자산회전률']
HK_columns = ['순이익증가율', '자기자본증가율', '1주당순이익']
HM_columns = ['재고자산 대 순운전자본비율']

# all_features의 컬럼을 업데이트
all_features.update(HS_features[HS_columns])
all_features.update(HK_features[HK_columns])
all_features.update(HM_features[HM_columns])

all_features.to_excel('코스피재무피쳐(최종).xlsx', index=False)
# 결과 확인
# print(all_features.head())
```
