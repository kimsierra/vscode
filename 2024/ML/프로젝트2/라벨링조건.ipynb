```python
# 파일을 불러옵니다 (엑셀 파일의 경우)
df = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/시장금리(일별)_20174908.xlsx')

# 데이터프레임의 컬럼명을 출력하여 올바른 컬럼명을 확인합니
```

```python
df
```

```python
df['날짜'] = pd.to_datetime(df['날짜'])
```

```python
df
```

```python
df['Year'] = df['날짜'].dt.year
df['Month'] = df['날짜'].dt.month

# 12월 데이터만 필터링합니다
december_df = df[df['Month'] == 12]

# 각 연도의 12월 마지막 데이터를 선택합니다
december_last = december_df.groupby('Year').apply(lambda x: x.loc[x['날짜'].idxmax()])
```
