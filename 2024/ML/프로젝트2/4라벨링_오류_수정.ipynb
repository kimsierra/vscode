## 코스피 라벨링 수정

```python
df = pd.read_csv('코스피_라벨링.csv', index_col = 0)
df = df.reset_index(drop=True)
```

```python
## 기업명이 중복되어서 잘못된부분이 있는지 확인
```

```python
df[df['회사명'] == '(주)신성이엔지']
```

```python
df[df['회사명'] == '삼성물산(주)']
```

```python
df[df['회사명'] == '한국제지(주)']
```

```python
## 한국제지[27970] 이자보상배율이 한국제지[2300]에 영향을 받아서 잘못 라벨링되어있음
df.iloc[[1195,1196],-1] = 0
df.iloc[[1195,1196],-4] = 0
```

```python
## 중간에 비어서 라벨링이 잘못된 데이터 확인
```

```python
df[df['회사명'] == '(주)유니켐']
## 중간에 비기전에 이미 부실로 라벨이 되어있어 채우지 않아도됨
```

```python
df.to_csv('코스피_라벨링_최종.csv')
```

## 코스닥 라벨링 수정

```python
df = pd.read_csv('코스닥_라벨링.csv', index_col = 0)
df = df.reset_index(drop=True)
```

### 동일 회사명에 다른 거래소코드로 인해 잘못라벨링된 데이터 수정

```python
# '(주)엔케이맥스[262760] 이 [182400]의 영향을 받아 부실로 잘못 라벨링됨
df.iloc[[13167,13168],-1] = 0
df.iloc[[13167,13168],-3] = 0
```

```python
df[df['회사명'] == '(주)엔케이맥스']
```

```python
df.iloc[[12828],-1] = 1
df[df['회사명'] == '(주)이큐셀']
```

```python
df.to_csv('코스닥_라벨링_최종.csv')
```
