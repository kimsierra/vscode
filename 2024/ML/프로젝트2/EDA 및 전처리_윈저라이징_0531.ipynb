### 윈저라이징

```python
# 윈저라이징 함수 정의
def winzorize_series(series):
    # 1%, 99% 백분위 수 계산
    lower_percentile = series.quantile(0.01)
    upper_percentile = series.quantile(0.99)
    
    # 극단값을 해당 백분위 값으로 대체
    series = series.clip(lower=lower_percentile, upper=upper_percentile)
    
    return series

# 숫자형 열에 대해 윈저라이징 수행
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_win = df.copy()
for col in numeric_cols[1:-1]:
    df_win[col] = winzorize_series(df[col])

# 결과 출력
df_win.describe()
```

```python
df.describe()
```

```python
df_win.to_csv("윈저라이징_inf제거.csv")
```

```python
# 윈저라이징 함수 정의
def winzorize_series(series):
    # 1%, 99% 백분위 수 계산
    lower_percentile = series.quantile(0.03)
    upper_percentile = series.quantile(0.97)
    
    # 극단값을 해당 백분위 값으로 대체
    series = series.clip(lower=lower_percentile, upper=upper_percentile)
    
    return series

# 숫자형 열에 대해 윈저라이징 수행
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_win2 = df.copy()
for col in numeric_cols[1:-1]:
    df_win2[col] = winzorize_series(df[col])

# 결과 출력
df_win2.describe()
```
