```python
import pandas as pd
import numpy as np
```

## 회사명에 거래소코드가 2개 이상인 기업 및 불연속적 데이터 확인

```python
df = pd.read_excel('코스피재무피쳐(최종).xlsx')
df.isna().sum()
```

### 코스피 동일 회사명에 거래소코드 2개이상 기업 확인

```python
df.groupby('회사명').filter(lambda x: x['거래소코드'].nunique() > 1)['회사명'].unique()
```

### 코스피 년도가 불연속적인 데이터 확인

```python
def is_non_consecutive(years):
    years = sorted([int(year.split('/')[0]) for year in years])
    return any(years[i] + 1 != years[i + 1] for i in range(len(years) - 1))

non_consecutive_companies = df.groupby('회사명').filter(lambda x: is_non_consecutive(x['회계년도']))

non_consecutive_companies['회사명'].unique()
```
