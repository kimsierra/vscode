```python
import pandas as pd
df1 = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스피효경연결피쳐.xlsx')
df2 = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스피효경개별피쳐.xlsx')
```

```python
common_columns = ['회사명', '거래소코드', '회계년도']

# 병합 (공통된 키를 기준으로 병합, 예를 들어 'ID' 열이 공통 열이라고 가정)
merged_df = pd.merge(df1, df2, on=common_columns, how='outer')

# '연결'이 포함된 열 이름만 선택
connected_columns = [col for col in merged_df.columns if '연결' in col]

# '연결'이 포함된 열을 기준으로 결측치 채우기
merged_df[connected_columns] = merged_df[connected_columns].fillna(method='ffill').fillna(method='bfill')

# '연결'이 포함되지 않은 열 제거
columns_to_drop = [col for col in merged_df.columns if '연결' not in col and col not in common_columns]
merged_df.drop(columns=columns_to_drop, inplace=True)

# # 결측치 채우기 (두 데이터프레임에서의 결측치를 서로 보완)
# merged_df.fillna(method='ffill', inplace=True)  # 앞의 값으로 채우기
# merged_df.fillna(method='bfill', inplace=True)  # 뒤의 값으로 채우기

merged_df.to_excel('코스피효경개별연결피쳐1.xlsx', index=False)
```

```python
strs = ['(*)(IFRS연결)', '(손실)(IFRS연결)','(손실)(IFRS연결)','[제조]', '(IFRS연결)','(결손금)(*)(IFRS연결)',
       '(결손금)','*']

for s in strs:

    # filterd_df.columns = filterd_df.columns.str.replace(s, "")
    merged_df.columns = merged_df.columns.str.replace(s, "")
    
merged_df.columns =merged_df.columns.str.strip()
merged_df.columns
```

```python
merged_df['회계년도'] = pd.to_datetime(merged_df['회계년도'])
merged_df.head(3)
```

```python
merged_df.isna().sum()
```

```python
# 전기매출액 구하기 (전기매출액은 현재 매출액의 한 칸 앞에 위치한 값)
merged_df['전기순이익'] = merged_df.groupby('회사명')['당기순이익'].shift(1)

# 매출액 증가율 계산 (분모에 절대값 적용)
merged_df['순이익증가율'] = ((merged_df['당기순이익'] - merged_df['전기순이익']) / abs(merged_df['전기순이익'])) * 100
# 새로운 파일로 저장
merged_df.to_excel('코스피효경개별연결그룹화(절대값).xlsx', index=False)
```

```python
# 데이터셋 불러오기
merged_df2 = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스피효경개별연결그룹화(절대값).xlsx')

# 전기말자산 구하기 (전기말자산은 현재 자산의 한 칸 앞에 위치한 값)
merged_df2['전기말 정상영업이익'] = merged_df2.groupby('회사명')['계속영업이익'].shift(1)
merged_df2['전기말 종업원수'] = merged_df2.groupby('회사명')['종업원수'].shift(1)
merged_df2['전기말 자본'] = merged_df2.groupby('회사명')['자본'].shift(1)    

# 자기자본증가율
merged_df2['자기자본증가율'] = ((merged_df2['자본'] / merged_df2['전기말 자본']) * 100) - 100

# 정상영업이익증가율
merged_df2['정상영업이익증가율'] = ((merged_df2['계속영업이익'] - merged_df2['전기말 정상영업이익']) / abs(merged_df2['전기순이익'])) * 100

# 기계투자효율
merged_df2['기계장치평균'] = (merged_df2['기계장치'] + merged_df2['기계장치'].shift(1)) / 2
merged_df2['기계투자효율'] = (merged_df2['부가가치'] / merged_df2['기계장치평균']) * 100

# 종업원수증가율
merged_df2['종업원수증가율'] = (merged_df2['종업원수'] - merged_df2['전기말 종업원수']) / abs(merged_df2['전기말 종업원수']) * 100

# 사내유보 대 자기자본비율
merged_df2['사내유보 대 자기자본비율'] = ((merged_df2['자본잉여금']+ merged_df2['이익잉여금']) / merged_df2['자본']) * 100

# 1주당순이익
merged_df2['1주당순이익'] = merged_df2['당기순이익'] / merged_df2['발행한 주식총수']

# 결과를 새로운 파일로 저장
merged_df2.to_excel('코스피효경개별연결그룹화(절대값전체피쳐).xlsx', index=False)
```
