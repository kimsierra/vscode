```python
print(len(np.where(df['이자보상배율(이자비용)(IFRS연결)']<0)[0]))
print(len(np.where(df['[제조]이자보상배율(이자비용)(IFRS)'] < 0)[0]))
```

```python
df['연결_영업이익'] = df['[제조]매출액(수익)(*)(IFRS연결)'] - df['[제조]매출원가(*)(IFRS연결)'] - df['[제조]판매비와 관리비(물류원가 등 포함)(*)(IFRS연결)']
df['개별_영업이익'] = df['[제조]매출액(수익)(*)(IFRS)'] - df['[제조]매출원가(*)(IFRS)'] - df['[제조]판매비와 관리비(물류원가 등 포함)(*)(IFRS)']
```

```python
df['연결_이자보상배율'] = df['연결_영업이익']/df['[제조]      이자비용(*)(IFRS연결)'] 
df['개별_이자보상배율'] = df['개별_영업이익']/df['[제조]      이자비용(*)(IFRS)'] 
```

```python
print(len(np.where(df['연결_이자보상배율']<0)[0]))
print(len(np.where(df['개별_이자보상배율'] < 0)[0]))
```

```python
print(df['이자보상배율(이자비용)(IFRS연결)'].isna().sum())
print(df['연결_이자보상배율'].isna().sum())
```

```python
def fill_missing(row):
    if pd.isna(row['이자보상배율(이자비용)(IFRS연결)']):
        return row['연결_이자보상배율']
    else:
        return row['이자보상배율(이자비용)(IFRS연결)']

df['이자보상배율(이자비용)(IFRS연결)'] = df.apply(lambda row: fill_missing(row), axis=1)
print(df['이자보상배율(이자비용)(IFRS연결)'].isna().sum())
```

```python
print(df['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'].isna().sum())
def fill_missing(row):
    if pd.isna(row['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)']):
        return row['[제조]영업활동으로 인한 현금흐름(간접법)(*)(IFRS)']
    else:
        return row['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)']

df['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] = df.apply(lambda row: fill_missing(row), axis=1)
print(df['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'].isna().sum())
```

```python
print(df['자본(*)(IFRS연결)(천원)'].isna().sum())
def fill_missing(row):
    if pd.isna(row['자본(*)(IFRS연결)(천원)']):
        return row['[제조]자본(*)(IFRS)']
    else:
        return row['자본(*)(IFRS연결)(천원)']

df['자본(*)(IFRS연결)(천원)'] = df.apply(lambda row: fill_missing(row), axis=1)
print(df['자본(*)(IFRS연결)(천원)'].isna().sum())
```

```python
ind1 = np.where(df['이자보상배율(이자비용)(IFRS연결)'] == np.inf)[0]
ind2 = np.where(df['이자보상배율(이자비용)(IFRS연결)'] == -np.inf)[0]
```

```python
no_inf_df = df.drop(index = ind1)
no_inf_df = no_inf_df.drop(index = ind2)
```

```python
no_inf_df.describe()
```

```python
no_inf_df = no_inf_df.reset_index(drop = True)
inf = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].max()
ind = np.where(no_inf_df['이자보상배율(이자비용)(IFRS연결)'] == inf)[0]
no_inf_df = no_inf_df.drop(index = ind)
no_inf_df.describe()
```

```python
max_value = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].max()
min_value = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].min()
```

```python
df.describe()
```

```python
df['이자보상배율(이자비용)(IFRS연결)'].replace(np.inf, max_value, inplace=True)
```

```python
df.describe()
```

```python
df['이자보상배율(이자비용)(IFRS연결)'].replace([-np.inf], min_value, inplace=True)
df.describe()
```

```python
df['이자보상배율(이자비용)(IFRS연결)']  = df['이자보상배율(이자비용)(IFRS연결)'].replace(df['이자보상배율(이자비용)(IFRS연결)'].max(),max_value)
```

```python
df.describe()
```

```python
df.to_csv('코스닥_라벨피쳐_전처리.csv')
```
