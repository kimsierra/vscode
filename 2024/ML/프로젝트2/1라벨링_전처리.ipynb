```python
df_kod['연결_영업이익'] = df_kod['[제조]매출액(수익)(*)(IFRS연결)'] - df_kod['[제조]매출원가(*)(IFRS연결)'] - df_kod['[제조]판매비와 관리비(물류원가 등 포함)(*)(IFRS연결)']
df_kod['개별_영업이익'] = df_kod['[제조]매출액(수익)(*)(IFRS)'] - df_kod['[제조]매출원가(*)(IFRS)'] - df_kod['[제조]판매비와 관리비(물류원가 등 포함)(*)(IFRS)']
```

```python
df_kod['연결_이자보상배율'] = df_kod['연결_영업이익']/df_kod['[제조]      이자비용(*)(IFRS연결)'] 
df_kod['개별_이자보상배율'] = df_kod['개별_영업이익']/df_kod['[제조]      이자비용(*)(IFRS)'] 
```

### 연결 결측값 개별 값으로 구해서 넣기

```python
def fill_missing(row):
    if pd.isna(row['이자보상배율(이자비용)(IFRS연결)']):
        return row['연결_이자보상배율']
    else:
        return row['이자보상배율(이자비용)(IFRS연결)']

df_kod['이자보상배율(이자비용)(IFRS연결)'] = df_kod.apply(lambda row: fill_missing(row), axis=1)
print(df_kod['이자보상배율(이자비용)(IFRS연결)'].isna().sum())
```

```python
def fill_missing(row):
    if pd.isna(row['이자보상배율(이자비용)(IFRS연결)']):
        return row['[제조]이자보상배율(이자비용)(IFRS)']
    else:
        return row['이자보상배율(이자비용)(IFRS연결)']

df_kod['이자보상배율(이자비용)(IFRS연결)'] = df_kod.apply(lambda row: fill_missing(row), axis=1)
print(df_kod['이자보상배율(이자비용)(IFRS연결)'].isna().sum())
```

```python
def fill_missing(row):
    if pd.isna(row['이자보상배율(이자비용)(IFRS연결)']):
        return row['개별_이자보상배율']
    else:
        return row['이자보상배율(이자비용)(IFRS연결)']

df_kod['이자보상배율(이자비용)(IFRS연결)'] = df_kod.apply(lambda row: fill_missing(row), axis=1)
print(df_kod['이자보상배율(이자비용)(IFRS연결)'].isna().sum())
```

```python
print(df_kod['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'].isna().sum())
def fill_missing(row):
    if pd.isna(row['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)']):
        return row['[제조]영업활동으로 인한 현금흐름(간접법)(*)(IFRS)']
    else:
        return row['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)']

df_kod['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] = df_kod.apply(lambda row: fill_missing(row), axis=1)
print(df_kod['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'].isna().sum())
```

```python
print(df_kod['자본(*)(IFRS연결)(천원)'].isna().sum())
def fill_missing(row):
    if pd.isna(row['자본(*)(IFRS연결)(천원)']):
        return row['[제조]자본(*)(IFRS)']
    else:
        return row['자본(*)(IFRS연결)(천원)']

df_kod['자본(*)(IFRS연결)(천원)'] = df_kod.apply(lambda row: fill_missing(row), axis=1)
print(df_kod['자본(*)(IFRS연결)(천원)'].isna().sum())
```

### 이자보상배율 무한대인값 처리하기 (음의 무한대 -> 최소값, 양의 무한대 ->최대값)

```python
ind1 = np.where(df_kod['이자보상배율(이자비용)(IFRS연결)'] == np.inf)[0]
ind2 = np.where(df_kod['이자보상배율(이자비용)(IFRS연결)'] == -np.inf)[0]
```

```python
no_inf_df = df_kod.drop(index = ind1)
no_inf_df = no_inf_df.drop(index = ind2)
```

```python
no_inf_df.describe()
```

```python
no_inf_df = no_inf_df.reset_index(drop = True)
inf = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].max()
ind = np.where(no_inf_df['이자보상배율(이자비용)(IFRS연결)'] == inf)[0]
no_inf_df = no_inf_df.drop(index = ind)
no_inf_df.describe()
```

```python
max_value = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].max()
min_value = no_inf_df['이자보상배율(이자비용)(IFRS연결)'].min()
```

```python
df_kod.describe()
```

```python
df_kod['이자보상배율(이자비용)(IFRS연결)'].replace(np.inf, max_value, inplace=True)
```

```python
df_kod.describe()
```

```python
df_kod['이자보상배율(이자비용)(IFRS연결)'].replace([-np.inf], min_value, inplace=True)
df_kod.describe()
```

```python
df_kod['이자보상배율(이자비용)(IFRS연결)']  = df_kod['이자보상배율(이자비용)(IFRS연결)'].replace(df_kod['이자보상배율(이자비용)(IFRS연결)'].max(),max_value)
```

```python
df_kod.describe()
```

```python
df_kod.to_csv('코스닥_라벨피쳐_전처리.csv')
```
