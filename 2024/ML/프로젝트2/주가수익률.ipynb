```python
import yfinance as yf
import pandas as pd

# Yahoo Finance에서 주식 데이터 가져오기
ticker = '^KS11'  # KOSPI 지수의 티커 심볼
data = yf.download(ticker, start='2010-01-01', end='2024-12-31', interval='1mo')

# 데이터프레임의 인덱스를 날짜 형식으로 변환
data.index = pd.to_datetime(data.index)

# 'Adj Close' 열만 사용
adj_close = data['Adj Close']

# 연도별 마지막 달의 종가를 추출
annual_close = adj_close.resample('Y').last()

# 연도별 주가 수익률 계산
annual_return = annual_close.pct_change() * 100

# NaN 값 제거
annual_return = annual_return.dropna()

# 연도별 주가 수익률을 보기 쉽게 데이터프레임으로 변환
annual_return_df = annual_return.to_frame(name='Annual Return (%)')

# 인덱스를 연도(YYYY) 형식의 문자열로 변환
annual_return_df.index = annual_return_df.index.year

# 엑셀 파일로 저장
excel_file_path = 'annual_return.xlsx'
annual_return_df.to_excel(excel_file_path, index_label='연도')
```

```python
import yfinance as yf
import pandas as pd

# Yahoo Finance에서 코스닥 지수 데이터 가져오기 (예: ^KQ11이 KOSDAQ 지수의 티커 심볼로 가정)
ticker = '^KQ11'
data = yf.download(ticker, start='2011-01-01', end='2022-12-31', interval='1mo')

# 데이터프레임의 인덱스를 날짜 형식으로 변환
data.index = pd.to_datetime(data.index)

# 'Adj Close' 열만 사용
adj_close = data['Adj Close']

# 연도별 마지막 달의 종가를 추출
annual_close = adj_close.resample('Y').last()

# 연도별 주가 수익률 계산
annual_return = annual_close.pct_change() * 100

# NaN 값 제거
annual_return = annual_return.dropna()

# 연도별 주가 수익률을 보기 쉽게 데이터프레임으로 변환
annual_return_df = annual_return.to_frame(name='Annual Return (%)')

# 인덱스를 연도(YYYY) 형식의 문자열로 변환
annual_return_df.index = annual_return_df.index.year

# 엑셀 파일로 저장
excel_file_path = 'kosdaq_annual_return.xlsx'
annual_return_df.to_excel(excel_file_path, index_label='연도')

print(f"연도별 주가 수익률 데이터가 '{excel_file_path}' 파일로 저장되었습니다.")
```

```python
import yfinance as yf
import pandas as pd

# 시작 및 종료 년도 설정
start_year = 2011
end_year = 2022

# 년도별 변동성을 저장할 데이터프레임 생성
volatility_df = pd.DataFrame(columns=['Year', 'Volatility'])

for year in range(start_year, end_year + 1):
    # 해당 년도의 데이터를 가져옴
    start_date = f"{year}-01-01"
    end_date = f"{year}-12-31"
    kospi_data = yf.download('^KS11', start=start_date, end=end_date)
    
    # 종가를 기준으로 일별 수익률을 계산
    kospi_data['Daily Returns'] = kospi_data['Close'].pct_change()
    
    # 변동성 계산 (일별 수익률의 표준편차)
    volatility = kospi_data['Daily Returns'].std()
    
    # 년도별 변동성을 데이터프레임에 추가
    volatility_df = volatility_df.append({'Year': year, 'Volatility': volatility}, ignore_index=True)

# 결과 출력
print(volatility_df)
```
