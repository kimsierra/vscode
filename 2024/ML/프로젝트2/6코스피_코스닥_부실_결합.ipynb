```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
kos = pd.read_csv('코스피재무피쳐_정리(최종).csv',index_col = 0)
kos_label = pd.read_csv('코스피_라벨링_최종.csv',index_col = 0)
```

### 코스피 재무 피쳐에 라벨링 결합

```python
kos_label = pd.read_csv('코스피_라벨링_최종.csv',index_col = 0)
final_kos = pd.merge(kos,kos_label[['회사명','거래소코드','회계년도','이자_부실','현금흐름_부실','자본_부실','부실']],how = 'left', on = ['회사명','거래소코드','회계년도'])
final_kos.head()
```

```python
final_kos.isna().sum()
```

```python
#코스피 결측치 46개 -> 상장 첫해년도로 인한 결측치 
ind = final_kos.iloc[:,3:8].isna().all(axis = 1)
ind = final_kos[ind].index
final_kos = final_kos.drop(index = ind).reset_index(drop=True)
final_kos.isna().sum()
```

```python
final_kos[final_kos['회사명'] == '대림통상(주)']
#2020년 데이터이므로 앞 값으로 채우기
final_kos= final_kos.fillna(method= 'ffill')
```

```python
final_kos.isna().sum()
```

## 코스피 무한대값 처리

```python
final_kos.isin([np.inf]).any()
```

```python
#무한대값 처리 -> 재고자산이 0인경우 무한대값이 나와 -> 재고자산이 0이면 재고자산의 회전률의 의미가 없어 0으로 처리
final_kos['재고자산회전률'].replace(np.inf, 0, inplace=True)
final_kos.isin([np.inf]).any().any()
```

```python
final_kos.sort_values(by = ['회사명','거래소코드','회계년도'],inplace= True)
final_kos.reset_index(drop=True, inplace = True)
```

```python
final_kos.to_csv('kos재무와라벨결합_부실처리x.csv')
final_kos.iloc[:,:53].to_csv('kos결측치처리된_재무.csv')
```

### 코스피 부실인기업 첫년도를 제외하고 제거

```python
risk_ind = np.where(final_kos['부실'] == 1)[0]
risk_df = final_kos.iloc[risk_ind,:]
risk_df = risk_df.drop_duplicates(subset=['회사명'])
risk_df = risk_df.reset_index(drop = True)
name = risk_df['회사명'].unique()
```

```python
for i in name:
    ind = np.where(final_kos['회사명'] == i)[0]
    final_kos = final_kos.drop(index = ind)
    final_kos = final_kos.reset_index(drop=True)
```

```python
final_kos['부실'].value_counts()
```

```python
risk_df['부실'].value_counts()
```

```python
final_kos_df = pd.concat([final_kos,risk_df],axis = 0)
final_kos_df.sort_values(by = ['회사명','거래소코드','회계년도'],inplace= True)
final_kos_df.reset_index(drop=True, inplace = True)
```

```python
final_kos_df.to_csv('코스피_라벨_결합_(최종).csv')
```

```python
final_kos_df['부실'].value_counts()
```

```python
final_kos_df.describe()
```

### 코스닥 재무 피쳐에 라벨링 결합

```python
kod = pd.read_csv('코스닥재무피쳐_정리(최종).csv',index_col = 0)
kod_label = pd.read_csv('코스닥_라벨링_최종.csv',index_col = 0)
final_kod = pd.merge(kod,kod_label[['회사명','거래소코드','회계년도','이자_부실','현금흐름_부실','자본_부실','부실']],how = 'left', on = ['회사명','거래소코드','회계년도'])
final_kod.head()
```

```python
#코스닥 결측치 중 179개 -> 상장 첫해년도로 인한 결측
final_kod.isna().sum()
```

```python
ind = final_kod.iloc[:,3:8].isna().all(axis = 1)
ind = final_kod[ind].index
final_kod = final_kod.drop(index = ind).reset_index(drop=True)
final_kod.isna().sum()
```

```python
#코스닥 나머지 194개 결측치인 행들 -> 결측치가 아닌 나머지값들도 0으로 들어감 -> 전체 행 결측치로 판단
```

```python
ind = final_kod[final_kod['부실'].isna()].index
final_kod = final_kod.drop(index = ind).reset_index(drop=True)
final_kod.isna().sum()
```

```python
##피에스케이는 상장첫년도 데이터로 제거
## 그외 씨제이프레시웨이,영진코퍼레이션은 앞의값으로 채움
final_kod = final_kod.drop(index = [12239]).reset_index(drop=True)
final_kod= final_kod.fillna(method= 'ffill')
final_kod.isna().sum()
```

## 코스닥 무한대값 처리

```python
final_kod.isin([np.inf]).any()
```

```python
#재고자산 회전률 무한대값 처리 -> 재고자산이 0인경우 무한대값이 나와 -> 재고자산이 0이면 재고자산의 회전률의 의미가 없어 0으로 처리
final_kod['재고자산회전률'].replace(np.inf, 0, inplace=True)
```

```python
#매출액증가율 무한대값 처리 ->  이전매출액이 0인 경우 무한대값 발생 -> 기준매출액이 0이면 증가율의 의미가 없어 0으로 처리
final_kod['매출액증가율'].replace(np.inf, 0, inplace=True)
```

```python
final_kod.isin([np.inf,-np.inf]).any().any()
```

```python
final_kod.sort_values(by = ['회사명','거래소코드','회계년도'],inplace= True)
final_kod.reset_index(drop=True, inplace = True)
```

```python
final_kod.to_csv('kod재무와라벨결합_부실처리x.csv')
```

```python
final_kod.iloc[:,:53].to_csv('kod_결측치처리된_재무.csv')
```

### 코스닥 부실 기업 첫년도를 제외하고 제거

```python
risk_ind = np.where(final_kod['부실'] == 1)[0]
risk_df = final_kod.iloc[risk_ind,:]
risk_df = risk_df.drop_duplicates(subset=['회사명'])
risk_df = risk_df.reset_index(drop = True)
name = risk_df['회사명'].unique()
```

```python
for i in name:
    ind = np.where(final_kod['회사명'] == i)[0]
    final_kod = final_kod.drop(index = ind)
    final_kod = final_kod.reset_index(drop=True)
```

```python
final_kod_df = pd.concat([final_kod,risk_df],axis = 0)
final_kod_df.sort_values(by = ['회사명','거래소코드','회계년도'],inplace= True)
final_kod_df.reset_index(drop=True, inplace = True)
```

```python
final_kod_df.to_csv('코스닥_라벨_결합_(최종).csv')
```
