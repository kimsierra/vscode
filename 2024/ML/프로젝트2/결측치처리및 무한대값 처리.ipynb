```python
#해외법인 날리기
df = df.drop(index = df[df['회사명'] == '차이나하오란리사이클링유한공사'].index)
df.reset_index(drop=True, inplace=True)
```

```python
df = df.drop(index = df[df['회사명'] == '에스앤씨엔진그룹리미티드'].index)
df.reset_index(drop=True, inplace=True)
```

```python
#무한대값 처리
df['재고자산회전률'].replace(np.inf, 0, inplace=True)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

fig = plt.figure(figsize=(25,35))
cnt = 1
for i in df.columns[3:-1]:
    ax = fig.add_subplot(10, 5, cnt)
    stats.probplot(df[i], dist=stats.hypsecant, plot=ax)
    ax.set_title(i, rotation=0)
    cnt = cnt + 1

plt.subplots_adjust(hspace=0.8)

plt.show()
```

```python
normal_enp = []
not_normal_enp = []
for column in df.columns[3:-1]:
    data = df[column].dropna()  
    stat, p = stats.shapiro(data)
    if p > 0.05:
        normal_enp.append(column)
    else:
        not_normal_enp.append(column)
```

```python
normal_enp
```

```python
not_normal_enp
```

```python
for i in not_normal_enp:
    df[i] = df[i].fillna(df[i].median())
```
