```python
strs = ['(IFRS연결)', '(*)(IFRS연결)(천원)','(*)(천원)','(손실)(천원)', '(간접법)(*)(IFRS연결)(천원)',
        '(수익)','(*)(IFRS연결)(천원)','(IFRS연결)(천원)','(손실)(IFRS연결)(천원)','(수익)(*)(IFRS연결)(천원)',
        '(*)(IFRS연결)','(천원)','(*)','(물류원가 등 포함)']

for s in strs:

    # filterd_df.columns = filterd_df.columns.str.replace(s, "")
    df.columns = df.columns.str.replace(s, "")
    
df.columns =df.columns.str.strip()
df.columns
```

```python
new_column_names = {'[제조]      매입채무 및 기타유동채무'.replace(" ", ""): '[제조]매입채무',
                    '[제조]      재고자산'.replace(" ", ""): '[제조]재고자산',
                    '영업활동으로 인한 현금흐름(간접법)': '영업현금흐름'}

# rename() 메서드를 사용하여 열 이름 변경
df = df.rename(columns=new_column_names)

# 변경된 열 이름 출력
print(df.columns)
```

```python
df['회계년도'] = pd.to_datetime(df['회계년도'])
df.head(3)
```

```python
# 전기매출액 구하기 (전기매출액은 현재 매출액의 한 칸 앞에 위치한 값)
df['전기매출액'] = df.groupby('회사명')['매출액'].shift(1)

# 매출액 증가율 계산 (분모에 절대값 적용)
df['매출액증가율'] = ((df['매출액'] - df['전기매출액']) / abs(df['전기매출액'])) * 100
# 새로운 파일로 저장
df.to_excel('코스닥그룹화(절대값).xlsx', index=False)
```

```python
# 데이터셋 불러오기
df = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스닥그룹화(절대값).xlsx')

# 전기말자산 구하기 (전기말자산은 현재 자산의 한 칸 앞에 위치한 값)
df['전기말자산'] = df.groupby('회사명')['자산'].shift(1)

# 총자본증가율 계산
df['총자본증가율(%)'] = ((df['자산'] - df['전기말자산']) / abs(df['전기말자산'])) * 100

# 영업현금흐름 대 차입금비율 계산
df['영업현금흐름 대 차입금비율'] = df['영업현금흐름'] / df['부채']

# 매출액총이익률 계산
df['매출액총이익률(%)'] = (df['매출총이익'] / df['매출액']) * 100

# 노동소득분배율 계산
# df['노동소득분배율(%)'] = (df['급여'] / df['부가가치(백만원)']) * 100 / 1000
df['노동소득분배율'] = df['[제조]인건비'] / (
    (df['매출액'] - df['[제조]매출원가'] - df['[제조]판매비와 관리비']) +
    df['[제조]인건비'] + df['[제조]   금융비용']) * 100 / 1000

# 매출채권 대 매입채무비율 계산
df['매출채권 대 매입채무비율'] = (df['[제조]      매출채권 및 기타유동채권'] / df['[제조]      매입채무 및 기타유동채무']) * 100

# 재고자산회전율 계산
df['재고자산회전율'] = df['매출액'] / ((df['[제조]      재고자산'] + df['[제조]      재고자산'].shift(1)) / 2)

# 결과를 새로운 파일로 저장
df.to_excel('코스닥그룹화(절대값전체피쳐).xlsx', index=False)
```
