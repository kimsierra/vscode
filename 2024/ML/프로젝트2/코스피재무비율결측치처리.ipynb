```python
import pandas as pd

# CSV 파일을 읽어옵니다. 파일 경로를 지정하세요.
df = pd.read_csv('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스피_재무비율_불표함_개별_연결_결측치제거_라벨링_정렬.csv')

# 결과를 저장할 리스트를 초기화합니다.
results = []

# 각 기업별로 첫 번째 '부실'에 '1'이 나오는 데이터를 찾습니다.
for company in df['회사명'].unique():
    company_data = df[df['회사명'] == company]
    first_one = company_data[company_data['부실'] == 1]
    
    # first_one이 비어 있지 않은 경우에만 추가
    if not first_one.empty:
        results.append(first_one.iloc[0])

# 리스트를 데이터프레임으로 변환합니다.
result_df = pd.DataFrame(results)

# 결과를 출력하거나 CSV 파일로 저장합니다.
print(result_df)

# 결과를 CSV 파일로 저장하려면 아래 주석을 해제하고 사용하세요.
# result_df.to_csv('path/to/save/result.csv', index=False)
```

```python
import pandas as pd

# CSV 파일을 읽어옵니다. 파일 경로를 지정하세요.
df = pd.read_csv('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/코스피_재무비율_불표함_개별_연결_결측치제거_라벨링_정렬.csv')

# 결과를 저장할 리스트를 초기화합니다.
results = []

# 각 기업별로 데이터를 처리합니다.
for company in df['회사명'].unique():
    company_data = df[df['회사명'] == company].reset_index(drop=True)
    first_one_idx = company_data[company_data['부실'] == 1].index
    
    if not first_one_idx.empty:
        first_one_idx = first_one_idx[0]  # 첫 번째 '1'의 인덱스
        # 앞, 뒤의 행을 제외하고 남은 데이터프레임 생성
        indices_to_keep = company_data.index.difference([first_one_idx - 1, first_one_idx, first_one_idx + 1])
        filtered_data = company_data.loc[indices_to_keep]
        results.append(filtered_data)

# 리스트를 데이터프레임으로 변환합니다.
if results:
    result_df = pd.concat(results)
else:
    result_df = pd.DataFrame()  # 비어있는 경우를 처리

# 결과를 출력하거나 CSV 파일로 저장합니다.
print(result_df)

# 결과를 CSV 파일로 저장합니다. 파일 경로를 지정하세요.
result_df.to_csv('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/결과.csv', index=False)
```
