```python
import pandas as pd
```

```python
df = pd.read_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/원화의 대미달러, 원화의 대위안_대엔 환율_20175130.xlsx')
df['날짜'] = pd.to_datetime(df['날짜'])
```

```python
df
```

```python
df['Year'] = df['날짜'].dt.year
df['Month'] = df['날짜'].dt.month

# 12월 데이터만 필터링합니다
december_df = df[df['Month'] == 12]
december_df2 = december_df.dropna()


# 각 연도의 12월 마지막 데이터를 선택합니다
december_last = december_df2.groupby('Year').apply(lambda x: x.loc[x['날짜'].idxmax()])
```

```python
december_last
```

```python
december_last.to_excel('C:/Users/SAMSUNG/Documents/vscode/2024/ML/프로젝트2/원화의 대미달러, 원화의 대위안_대엔 환율_20175130.xlsx', index=False)
```
