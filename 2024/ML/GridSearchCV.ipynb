#### sklearn.model_selection.GridSearchCV

- class sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, refit=True, cv=None, verbose=0(수다쟁이, 자세히 설명), pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

iris = load_iris()
dt_clf = DecisionTreeClassifier(random_state=112)
train_data = iris.data
train_label = iris.target


# DecisionTree
param_dt = {"max_depth" : [2,5,7],
          "min_samples_split" : [2,3]
          }

gscv_dt = GridSearchCV (estimator = dt_clf, param_grid = param_dt, scoring ='accuracy', cv = 3, refit=True, n_jobs=1, verbose=0)

gscv_dt.fit(train_data,train_label)
```
